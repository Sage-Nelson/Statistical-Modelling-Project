# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import requests
import pandas as pd
import json

# Get the Paris network info from CityBikes
citybikes_url = "http://api.citybik.es/v2/networks"
response = requests.get(citybikes_url)
response
data = response.json()
# data
city_name = "Paris"
city_networks = [n for n in data['networks'] if city_name.lower() in n['location']['city'].lower()]
city_networks
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
for net in city_networks:
    print(f"Name: {net['name']}, ID: {net['id']}")
network_id = "velib"
url = f"https://api.citybik.es/v2/networks/{network_id}"
response = requests.get(url)
response
network_data = response.json()
network_data
```

Put your parsed results into a DataFrame.

```python
stations = network_data['network']['stations']
df_stations = pd.DataFrame(stations)
df_stations = df_stations[['name', 'latitude', 'longitude', 'free_bikes', 'empty_slots']]
df_stations
```

```python
df_stations.to_csv('df_stations.csv', index=False)
```
